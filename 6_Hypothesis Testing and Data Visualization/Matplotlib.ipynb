# Matplotlib Examples

```
import matplotlib.pyplot as plt
```

```
X = [1, 2, 3]
Y = [1, 4, 9]
Z = [1, 8, 27]

plt.plot(X, Y, label='Squared', color='c')
plt.plot(X, Z, label='Cubed', color='m')

plt.xlabel('independent values')
plt.ylabel('dependent values')
plt.title('Important plot \n check it out')
plt.legend()
plt.show()
```

```
plt.scatter(X, Y, label='Squared', color='c', marker='*', s=1000)
plt.scatter(X, Z, label='Cubed', color='m')

plt.xlabel('independent values')
plt.ylabel('dependent values')
plt.title('Important plot \n check it out')
plt.legend()
plt.show()
```

```
plt.bar(X, Y, label='Squared', color='c')
plt.bar(X, Z, label='Cubed', color='m')

plt.xlabel('independent values')
plt.ylabel('dependent values')
plt.title('Important plot \n check it out')
plt.legend()
plt.show()
```

```
plt.hist(Y, 2, width=0.8, label='Squared', color='c')
plt.hist(Z, 2, width=0.8, label='Cubed', color='m')

plt.xlabel('independent values')
plt.ylabel('dependent values')
plt.title('Important plot \n check it out')
plt.legend()
plt.show()
```

```
values = [8, 8, 4, 4]
labels = ['sleeping', 'studying', 'playing', 'eating']
          
plt.pie(values, labels=labels, startangle=90, shadow=1, autopct='%1.1f', explode=(0, 0, 0.2, 0))

plt.title('Important plot \n check it out')
plt.show()          
```

```
values1 = [8, 8, 4, 4]
values2 = [8, 4, 8, 8]
values3 = [4, 8, 8, 4]
values4 = [4, 4, 4, 8]
x = [1, 2, 3, 4]
plt.stackplot(x, values1, values2, values3, values4)

plt.title('Important plot \n check it out')
plt.show() 
```
